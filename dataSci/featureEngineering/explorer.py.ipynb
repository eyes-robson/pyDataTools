```python
import multiprocessing as mp
import numpy as np
import itertools as it
from sklearn import preprocessing as pre
from sklearn.datasets import load_diabetes
```

```python
X,y,__,feats = tuple(x for x in load_diabetes().values())
display(X)
display(y)
```

```python
import pandas as pd
X = pd.DataFrame(X)
X.columns = feats
display(X.head())
```

```python
fe_polynomial = pre.PolynomialFeatures(degree=3)
display(X.head())
fe_polynomial.fit(X);
df = pd.DataFrame(fe_polynomial.transform(X))
df.head()
```

```python
fe_polynomial = pre.PolynomialFeatures(degree=2,include_bias=False)
display(X.head())
fe_polynomial.fit(X);
df = pd.DataFrame(fe_polynomial.transform(X))
df.head().iloc[:,0:15]
```

```python
fe_polynomial = pre.PolynomialFeatures(degree=2,include_bias=False)
df = pd.DataFrame(fe_polynomial.fit_transform(X))
df.head().iloc[:,0:15]
```
